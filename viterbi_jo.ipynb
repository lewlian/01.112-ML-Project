```python
def inner_viterbi(sentence, em_params, tr_params, tags):
#     words = file.readlines()
#     words = [x for x in sentence if x != '\n']
    words = sentence
    
    # FORWARD
    n = len(words)

    T= [[0 for i in range(len(tags))] for j in range(n) ]

    arg = [[0 for i in range(len(tags))] for j in range(n) ]
    
#     print(words)
    
    # base case/ first layer
    for u in range(len(tags)):
        word = words[0].rstrip("\n")
#         print(word)
        try:
#             T[0][u] = np.log(tr_params[tags[u]]["START"]*em.getLikelihood(em_params, words[0], tags[u]))
            a = tr_params[tags[u]]["START"]
            b = em.getLikelihood(em_params, word[0], tags[u],w)
#             print('viterbi base:',a,b)
            if a!= 0 and b!= 0:
                T[0][u] = np.log(a*b)
            else:
                T[0][u] = 0
        except KeyError:
            T[0][u] = 0
        except TypeError:
            T[0][u] = 0
    
#     nif = np.log(0)
    for k in range(1, n):
#         print(k)
        word = words[k].rstrip("\n")
#         print(word)
        for v in range(len(tags)):
            find_max = []
            find_arg = []
            
            for u in range(len(tags)):     
                try:
                    pilog = T[k-1][u]
                    pi = np.exp(pilog)
                    if(tr_params[tags[v]] != 0) :
                        a = tr_params[tags[v]][tags[u]]
                        b = em.getLikelihood(em_params, word, tags[v], w)
#                         alog = np.log(a)
#                         blog = np.log(b)
#                         print('viterbi:',a,b,tags[v])
                        if pilog!=0 and a!=0 and b!=0:
#                             print('Say yes T!')
                            value = np.log(pi*a*b)
                            arg_value = np.log(pi*a)
                        else:
#                             print('something is zero')
                            value = 0
                            arg_value=0
                    else: 
                        value = 0
                        arg_value = 0
                except KeyError:
                    value = 0
                    arg_value = 0


                find_max.append(value)

                find_arg.append(arg_value)
            
            tmax = -sys.maxsize
            for i in range(len(find_max)):
                value = find_max[i]
                if value!=0:
                    if value>tmax:
                        tmax = value
            if tmax == -sys.maxsize:
                T[k][v] = 0
            else:
                T[k][v] = tmax
                
            argmax = -sys.maxsize
            for i in range(len(find_max)):
                value = find_arg[i]
                if value!=0:
                    if value>argmax:
                        argmax = value
            if argmax == -sys.maxsize:
                arg[k][v] = 0
            else:
                arg[k][v] = find_arg.index(argmax)
#             if find_max == []:
#                 print('empty find_max')
#                 T[k][v] = 0
#             else:
#                 print('added max to T')
#                 T[k][v] = max(find_max)
#             if find_arg == []:
#                 print('empty find_arg')
#                 arg[k][v] = index_unknown
#             else:
#                 print('added arg(max) to arg')
#                 print(find_arg)
#                 arg[k][v] = find_arg.index(max(find_arg))


    #BACKWARD
    index = [0]*n
    y = [0]*n
    
    #last value
    true_max = -sys.maxsize
    for m in range(len(T[n-1])):
        value = T[n-1][m]
        if value != 0:
            if value > true_max:
                true_max = value
    if true_max == -sys.maxsize:
        index[n-1] = index_unknown
        y[n-1] = tags[index[n-1]]
    else:
        index[n-1] = T[n-1].index(true_max)
        y[n-1] = tags[index[n-1]]
    
    #the rest of values using stored arg table
    for j in range(n-1,0,-1):
#         print('find rest of values:',j)
        index[j-1] = arg[j][index[j]]
#         print('prev',index[j])
#         print('now',index[j-1])
        y[j-1] = tags[index[j-1]]
    
    return y,T,arg
```

```python
def viterbi(file, em_params, tr_params, tags):
    file_list = file.readlines()
    sentences = [list(group) for k, group in groupby(file_list, lambda x: x == "\n") if not k]
    total_y = []
    for i in range(len(sentences)):
        print(i)
        sentence = sentences[i]
        y,T,arg = inner_viterbi(sentence, em_params, tr_params, tags)
        total_y += y
        total_y += [0]
    return total_y
```

```python
def predictions_file(inputFile, outputfile, y):
    f = open(outputfile, 'w+', encoding="utf8")

    finput = open(inputFile, 'r', encoding="utf8")
    lines = finput.readlines()
    print(len(lines))
    print(len(y))
    for i in range(len(lines)):
        if (lines[i].rstrip() != ""):
            word = lines[i].rstrip()
            # print(word, y[i+1])
            if (word == '\n'):
                f.write(word)
            else:
                f.write(word+" "+y[i]+"\n")
            
        else:
            f.write('\n')
    f.close()
    print('Finished writing to File.')
```
